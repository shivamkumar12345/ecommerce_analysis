## Ecommerce Sales Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
pd.read_excel("../Superstore.xlsx")
# print(csv)
```

```python
csv= pd.read_excel("../Superstore.xlsx")

```

```python
csv.isnull().sum()
```

```python
csv.describe()
```

```python
csv.sort_values(['Profit','Sales'],axis=0,ascending=[False,False],inplace=True)
```

```python
fig= px.bar(csv.head(20),x='Region', y="Sales",color='Category',width=700)
fig.show()
```

```python
fig= px.bar(csv.head(50),x='City', y="Sales",color='Profit',hover_data=['Category'],width=700,title="Sales in region with highest profit")
fig.show()
```

```python
fig =px.bar(csv.head(100),x='Sub-Category',color='Category',width=700,title="Order Count of Categories Sub-Catagory")
fig.show()
```

```python
fig =px.bar(csv.head(20),x='Customer Name',y='Quantity',color='Profit',hover_data=['Discount','Category'],width=700,height=600,title="Number of Order By Customer")
fig.show()
```

```python
fig =px.pie(csv,names='Sub-Category',values='Profit',title="Order Count of Categories Sub-Catagory")
fig.show()
```

```python
csv.sort_values(['Sales','Profit'],axis=0,ascending =[False,False],inplace=True)
```
